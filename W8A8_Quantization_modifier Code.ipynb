```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from llmcompressor.modifiers.quantization import QuantizationModifier
from llmcompressor.modifiers.smoothquant import SmoothQuantModifier
from llmcompressor.transformers import oneshot
from datasets import load_dataset
```

```python
import torch
model_stub = "deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B"
model_name = model_stub.split("/")[-1]

num_samples = 128
max_seq_len = 1024

tokenizer = AutoTokenizer.from_pretrained(model_stub)

model = AutoModelForCausalLM.from_pretrained(
    model_stub,
    device_map="auto",
    torch_dtype=torch.float32
)
```

```python
print(model.base_model.layers[0].mlp.gate_proj.weight.dtype)
```

```python
def preprocess_fn(example):
  return {"text": tokenizer.apply_chat_template(
      example["messages"],
      add_generation_prompt=False,
      tokenize=False
  )}
ds = load_dataset("neuralmagic/LLM_compression_calibration", split="train")
ds = ds.map(preprocess_fn)
```

```python
QuantizationModifier(
    targets="Linear",
    scheme="W8A8",
    ignore=["lm_head"],
    dampening_frac=0.1,
)

recipe = [
    SmoothQuantModifier(smoothing_strength=0.9),
    QuantizationModifier(
        targets="Linear",
        scheme="W8A8",
        ignore=["lm_head"],
        dampening_frac=0.1,
    ),
]
oneshot(
    model=model,
    dataset=ds,
    recipe=recipe,
    max_seq_length=max_seq_len,
    num_calibration_samples=num_samples,
)

save_path = "/content/drive/MyDrive/CS594/" + model_name + "-quantized.w8a8"
model.save_pretrained(save_path)
tokenizer.save_pretrained(save_path)
print(f"Model and tokenizer saved to: {save_path}")
```
